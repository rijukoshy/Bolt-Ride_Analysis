### solve the problem of missing values

### solve start and end date and time converting to datetime


```python
bolt['START_DATE'] = pd.to_datetime(bolt['START_DATE'], errors='coerce')
bolt['END_DATE'] = pd.to_datetime(bolt['END_DATE'], errors='coerce')
```

```python
bolt.head(10)
```

### seperate the date and time and month and year


```python
# Convert to datetime if 'START_DATE' exists
if 'START_DATE' in bolt.columns:
    bolt['START_DATE'] = pd.to_datetime(bolt['START_DATE'], errors='coerce')
    bolt['END_DATE'] = pd.to_datetime(bolt['END_DATE'], errors='coerce')

    bolt['start_date'] = bolt['START_DATE'].dt.date
    bolt['start_time'] = bolt['START_DATE'].dt.time
    bolt['end_date'] = bolt['END_DATE'].dt.date
    bolt['end_time'] = bolt['END_DATE'].dt.time

    # Extract month and year from 'start_date'
    bolt['month'] = bolt['START_DATE'].dt.month
    bolt['year'] = bolt['START_DATE'].dt.year
else:
    print("Warning: 'START_DATE' or 'END_DATE' missing. Adding empty columns.")
    
    # Add missing columns with NaN
    bolt['start_date'] = pd.NA
    bolt['start_time'] = pd.NA
    bolt['end_date'] = pd.NA
    bolt['end_time'] = pd.NA
    bolt['month'] = pd.NA
    bolt['year'] = pd.NA

```

```python
bolt.head(10)
```

### remove START_DATE END_DATE and rearrange the columns


```python
bolt.drop(['START_DATE','END_DATE'],axis=1,inplace=True)
bolt = bolt[['start_date','start_time','end_date','end_time','month','year','CATEGORY','START','STOP','MILES','PURPOSE']]
```

```python
bolt.month.value_counts()
```

```python
bolt.month.isnull().sum()
```

### convert month to integer


```python
# Bolt ride.month.fillna(fill,inplace=True)
bolt['month'] = bolt['month'].fillna(bolt['month'].mode()[0])
bolt['month'] = bolt['month'].astype(int)
```

```python
bolt.head(10)
```

```python
bolt.month.count()
```

```python
bolt.month.value_counts()
```

### convert year to integer


```python
bolt['year'] = bolt['year'].fillna(bolt['year'].mode()[0])
bolt['year'] = bolt['year'].astype(int)
```

```python
bolt.dtypes
```

```python
# convert start_time and end_time to hours and minutes
bolt['start_hour'] = bolt['start_time'].apply(lambda x: x.hour)
bolt['start_minute'] = bolt['start_time'].apply(lambda x: x.minute)
bolt['end_hour'] = bolt['end_time'].apply(lambda x: x.hour)
bolt['end_minute'] = bolt['end_time'].apply(lambda x: x.minute)
```

```python
# rearrange the columns and drop start_time and end_time columns
bolt = bolt[['start_date','start_hour','start_minute','end_date','end_hour','end_minute','month','year','CATEGORY','START','STOP','MILES','PURPOSE']]
```

```python
# create a new column for duration of the trip to be in minutes
bolt['duration'] = bolt['end_hour']*60 + bolt['end_minute'] - (bolt['start_hour']*60 + bolt['start_minute'])

```

```python
# convert miles to km
bolt['MILES'] = bolt['MILES']*1.60934
#calculate speed
bolt['speed'] = bolt['MILES']/(bolt['duration']/60)
```
